```python
# Initial imports
import os
import requests
import json
import pandas as pd
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi

%matplotlib inline
```

```python
# Load .env enviroment variables
load_dotenv()
```

```python
# Set Alpaca API key and secret key
alpaca_api_key = os.getenv('ALPACA_API_KEY')
alpaca_secret_key = os.getenv('ALPACA_SECRET_KEY')

# Create the Alpaca API object
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version='v2')
```

```python
# Format current date as ISO format
election_day = pd.Timestamp('2020-11-03', tz='America/Chicago').isoformat()

# Set the tickers for Energy, Financial, Tech, and Healthcare 
# per https://stockcharts.com/freecharts/sectorsummary.html?O=6#
tickers = ['XLE', 'XLF', 'XLK', 'XLV']

# Set timeframe to '1D' for Alpaca API
timeframe = "1D"

# Get current closing prices for SPY and AGG
# (use a limit=1000 parameter to call the most recent 1000 days of data)
election_2020_prices_df = alpaca.get_barset(
    tickers,
    timeframe,
    start = election_day,
    end = election_day,
    limit = 1000
).df

# Preview DataFrame
election_2020_prices_df
```
